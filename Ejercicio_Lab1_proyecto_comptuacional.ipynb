```python
import matplotlib.pyplot as plt
import numpy as np

# Supongamos que tienes una imagen cargada en la variable "imagen"

# Verify the loaded image shape
print("Image shape:", imagen.shape)

# Get the dimensions of the image
filas, columnas, capas = imagen.shape
print("Number of rows:", filas)

# Calculate the middle row and middle column
fi = filas // 2
co = columnas // 2
print("Middle row index:", fi)
line_width = 10  # Ancho de la línea

# Aumentar el ancho de la línea horizontal y vertical
for i in range(columnas):
    imagen[fi - (line_width // 2):fi + (line_width // 2), i] = [255, 255, 0]
for i in range(filas):
    imagen[i, co - (line_width // 2):co + (line_width // 2)] = [255, 255, 0]

# Aumentar el grosor de la diagonal
for t in range(filas):
    for j in range(columnas):
        if t == j or t == filas - j - 1:
            imagen[t, j] = [255, 255, 0]

# Ancho del margen
margin_width = 20

# Color del margen (amarillo en RGB)
margin_color = [255, 255, 0]

# Crear una nueva imagen con el margen
imagen_con_margen = np.copy(imagen)

# Agregar el margen en la parte superior e inferior
imagen_con_margen[:margin_width, :, :] = margin_color
imagen_con_margen[-margin_width:, :, :] = margin_color

# Agregar el margen en los lados izquierdo y derecho
imagen_con_margen[:, :margin_width, :] = margin_color
imagen_con_margen[:, -margin_width:, :] = margin_color

# Display the modified image with margin
plt.imshow(imagen_con_margen)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Supongamos que tienes una imagen cargada en la variable "imagen"

# Calcular el ancho de la línea y el margen
line_width = 10
margin_width = 20

# Color de la línea y el margen (amarillo en RGB)
line_color = [255, 255, 0]
margin_color = [255, 255, 0]

# Aumentar el ancho de la línea horizontal y vertical
imagen[imagen.shape[0] // 2 - line_width // 2:imagen.shape[0] // 2 + line_width // 2, :] = line_color
imagen[:, imagen.shape[1] // 2 - line_width // 2:imagen.shape[1] // 2 + line_width // 2] = line_color

# Aumentar el grosor de la diagonal
imagen[range(imagen.shape[0]), range(imagen.shape[1])] = line_color
imagen[range(imagen.shape[0]), range(imagen.shape[1]-1, -1, -1)] = line_color

# Crear una nueva imagen con el margen
imagen_con_margen = np.copy(imagen)

# Agregar el margen
imagen_con_margen[:margin_width, :, :] = margin_color
imagen_con_margen[-margin_width:, :, :] = margin_color
imagen_con_margen[:, :margin_width, :] = margin_color
imagen_con_margen[:, -margin_width:, :] = margin_color

# Mostrar la imagen modificada con margen
plt.imshow(imagen_con_margen)
plt.show()
```
