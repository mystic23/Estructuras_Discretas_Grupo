```python
import cv2, numpy as np,matplotlib.pyplot as plt
T=cv2.imread("escudo junior.jpg");plt.imshow(T)
filas,columnas,capas = T.shape;print(T.shape)
b=int(input("digita grosor de la linea. b="))
T[filas//2-b:filas//2+b,:,:]=0
T[:,columnas//2-b:columnas//2+b,:]=0;plt.imshow(T)
for f in range(filas): T[f,f,:]=0;
for f in range(filas): T[f, np.abs(filas-columnas)-f,:]=4;
plt.imshow(T)
```
