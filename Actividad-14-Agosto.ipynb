##Formula:

1. Tenemos que la función recurrente:
  $$ f_{n} = (n-2) +f(n-1) ; f(3)=0 ; n\ge 4$$

2. Tenemos que la función no recurrente es:

  $$f_{n} = \frac{n \cdot (n-3) }{2} ; n\ge 3 $$

3. Tenemos los numéros de catalan:

  $$ \frac{2n!}{(n+1)!n!}$$



```python
import numpy as np
import scipy as sc
import sympy as sp
import matplotlib.pyplot as plt

# n = int(input("Ingrese un numero positivo n>=4 n =  "))
n = 12
def func(n):
  if n == 3:
    return 0
  else:
    return (n-2)+func(n-1)

v = np.zeros(n+3)
for i in range(4,n+3):
  v[i]=func(i)

plt.plot(v)
plt.title(f' (n-2)+f(n-1)')
plt.xlabel("Eje x")   # Establece el título del eje x
plt.ylabel("Eje Y")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def func(n):
    return int(n * (n - 3) / 2)

n = 12

v1 = np.zeros(n + 3)

for i in range(3, n + 1):
    diagonales = func(i)
    v1[i] = diagonales
    print(i, " contiene ", diagonales, " diagonales interiores")

# Graficar la función
plt.plot(np.arange(3, n + 1), v1[3:n + 1], linestyle='-', color='b', label='Diagonales interiores')
plt.xlabel('Número de lados (n)')
plt.ylabel('Número de diagonales interiores')
plt.title('Diagonales interiores en función del número de lados')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_fraction(n):
    return 2 * np.math.factorial(n) / (np.math.factorial(n + 1) * np.math.factorial(n))

# Calcular la función para diferentes valores de 'n'
n_values = np.arange(1, 21)  # Valores de n desde 1 hasta 20
results = [calculate_fraction(n) for n in n_values]

# Graficar los resultados
plt.plot(n_values, results, linestyle='-', color='b')
plt.xlabel('Valor de n')
plt.ylabel('Resultado')
plt.title('Gráfico de la función $\\frac{2n!}{(n+1)!n!}$')
plt.grid(True)
plt.show()
```

##El triangulo de pascal
Pascal lo resolvio de forma empirica con el problema del binomio, donde:
$(x+r)^n = \sum_{k=0}^{n} \begin{pmatrix} n\\k \end{pmatrix} x^{n-k} \cdot y^n $

image.png

Entonces tenemos que:
$\begin{pmatrix} n\\n-k \end{pmatrix}=\begin{pmatrix} n\\k \end{pmatrix}= \frac{n!}{k!(n-k)!}, donde k <= n $

Seguidamente tenemos que:
$\begin{pmatrix} n\\k \end{pmatrix}=\begin{pmatrix} n-1\\k \end{pmatrix}+\begin{pmatrix} n-1\\k-1 \end{pmatrix}  $


```python
import numpy as np
#pendiente 1 es secundaria
#pendiente -1 es principales
n = 4
nf = n + 1
nc = 2 * n + 1
tp = np.zeros((nf, nc))

# Coloco el 1 centrado
tp[0, nc // 2] = 1
# dp = j-1 #diagonal principal
# ds = j+1 #diagonal secundaria
# Llenar las diagonales
for i in range(1, nf):
    tp[i, n-i] = 1  # Diagonal secundaria izquierda
    tp[i, n+i] = 1
    for j in range(n-i,n+i,2):
      tp[i,j] = tp[i-1,j-1]+tp[i-1,j+1] # Diagonal secundaria derecha
print("\n".join(["".join(['{:4}'.format(int(item)) if item != 0 else '    ' for item in row]) for row in tp]))

```
